```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rc('text', usetex=True)
mpl.rc('font', family='serif', weight = 'normal', style = 'normal',size = '18')

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 18,
        }

# make these smaller to increase the resolution
dx, dy = 0.10, 0.005

#reading data
time1, nucley1 = np.loadtxt("decay_0,05.dat",unpack=True)
time2, nucley2 = np.loadtxt("decay_0,2.dat",unpack=True)

#customizando o gráfico

fig1 = plt.figure(figsize = (7,5))
plt.subplots_adjust(top=0.96, bottom=0.14, left=0.14, right=0.98, wspace=0.5, hspace=0.39)
#subplot(nrows, ncols, index, **kwargs)
ax1 = fig1.add_subplot(1,1,1)


ax1.grid(b=True, which='major', color='DarkGray', alpha=0.4, linestyle=':', linewidth=2)
ax1.set_xlim([0.0, 5.0])
ax1.set_ylim([0.0, 110])
ax1.set_ylabel("$\\textnormal{Number of nuclei}$ ",fontsize=18)
ax1.set_xlabel('$t \ (s)$',fontsize=25)
ax1.plot(time1,nucley1,'.-', color='Blue', lw=1, ms=5, markevery=1, label='$\Delta t=0.05$')
ax1.plot(time2, nucley2,'.-', color='red', lw=1, ms=5, markevery=1, label='$\Delta t=0.2$')

legend = ax1.legend(loc='best', shadow=False, fontsize='18')

plt.show()
```
